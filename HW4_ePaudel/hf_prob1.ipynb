```
#population parameter for uniform distribution
a = 0
b = 1
mean = (a+b)/2
variance = 1/12*(b-a)**2
skewness = 0
kurtosis = -(6/5)
```

```
#to generate random numbers in uniform distribution
np.random.seed(212)
dist = {}
sample_mean = np.array([])
sample_var = np.array([])
sample_skew = np.array([])
sample_kurt = np.array([])
k = [4**n for n in np.arange(1,11)]
center = 0
for i in np.arange(0,10):
  dist[i] = sp.stats.uniform.rvs(size=k[i], loc=center, scale=1)
  sample_mean = np.append(sample_mean,dist[i].mean())
  sample_var = np.append(sample_var,np.var(dist[i]))
  sample_skew = np.append(sample_skew,sp.stats.skew(dist[i]))
  sample_kurt = np.append(sample_kurt,sp.stats.kurtosis(dist[i]))
```

```
log2k = np.log2(asarray(k))
```

```
#plotting mean of samples of different size
ax=pl.figure(figsize=(10,6)).add_subplot(111)
ax.plot(log2k , sample_mean,'o',label='sample mean')
ax.set_xlabel('$log_{2}$[sample size]',fontsize = 18)
ax.set_ylabel('mean',fontsize = 18)
ax.plot([min(log2k), max(log2k)], [mean, mean], 'k',label = 'expected mean')
pl.legend(loc='lower right', fontsize = 18)
pl.show()
```

###Figure 1: Mean of samples of different sizes plotted against $log_{2}$ of sample size. As sample size increases, sample mean converges to population mean. 

```
ax=pl.figure(figsize=(10,6)).add_subplot(111)
ax.plot(log2k , sample_var,'o',label='sample variance')
ax.set_xlabel('$log_{2}$[sample size]',fontsize = 18)
ax.set_ylabel('variance',fontsize = 18)
ax.plot([min(log2k), max(log2k)], [variance, variance], 'k',label = 'expected variance')
pl.legend(loc='upper right', fontsize = 18)
pl.show()
```

###Figure 2: Variance of samples of different sizes plotted against $log_{2}$ of sample size. As sample size increases, sample variance converges to population variance. 

```
ax=pl.figure(figsize=(10,6)).add_subplot(111)
ax.plot(log2k , sample_skew,'o',label='sample skewness')
ax.set_xlabel('$log_{2}$[sample size]',fontsize = 18)
ax.set_ylabel('skewness',fontsize = 18)
ax.plot([min(log2k), max(log2k)], [skewness, skewness], 'k',label = 'expected skewness')
pl.legend(loc='upper right', fontsize = 18)
pl.show()
```

###Figure 3: Skewness of samples of different sizes plotted against $log_{2}$ of sample size. As sample size increases, skewness of randomly generated sample data converges to population skewness 0. 

```
ax=pl.figure(figsize=(10,6)).add_subplot(111)
ax.plot(log2k , sample_kurt,'o',label='sample kurtosis')
ax.set_xlabel('$log_{2}$[sample size]',fontsize = 18)
ax.set_ylabel('kurtosis',fontsize = 18)
ax.plot([min(log2k), max(log2k)], [kurtosis, kurtosis], 'k',label = 'expected kurtosis')
pl.legend(loc='upper right', fontsize = 18)
pl.show()
```

###Figure 4: Kurtosis of samples of different sizes plotted against $log_{2}$ of sample size. As sample size increases, kurtosis of randomly generated sample converges to population kurtosis -1.2. 

###Conclusion:  
When we study the plots, we can conclude that after the sample size is large (< 1000), various moments like mean, variance, skewness and kurtosis of randomly generated sample agrees with that of distribution from which parameters are taken. But when the sample size is small, those values fluctuates a lot.
